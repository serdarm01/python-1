```python
def faktoryelHesaplama(a):
  fakto=1
  for i in range(1,a+1):
    fakto=i*fakto
  return fakto

hesap=int(input("Faktoryeli hesaplanacak hesap:"))
faktoryelHesaplama(hesap)
```
